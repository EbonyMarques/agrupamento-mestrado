Aqui ocorrem:
1. a eliminação de atributos sem variância de dados ("VarianceThreshold" com threshold = 0);
2. a eliminação de atributos desnecessários (códigos de área do CINE Brasil);

```python
import pandas as pd
from pandas_profiling import ProfileReport
from sklearn.feature_selection import VarianceThreshold
```

```python
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('cursos-ufrpe.csv',encoding='ISO-8859-1',sep=';')
```

```python
df
```

```python
df.shape
```

```python
sel = VarianceThreshold(threshold=(0))
```

```python
df1 = df.drop(['no_curso','no_municipio','no_cine_area_geral','tp_grau_academico', 'no_regiao', 'no_uf',
               'sg_uf', 'no_cine_rotulo', 'co_cine_rotulo', 'no_cine_area_especifica', 'no_cine_area_detalhada',
               'enade_faixa', 'cpc_faixa'],axis=1)
```

```python
df_selected = sel.fit_transform(df1)
cols_selected = df1.columns[sel.get_support(indices=True)]
df_selected = pd.DataFrame(df_selected, columns=cols_selected)
```

```python
df_selected
```

```python
df_selected.shape
```

```python
df_selected = df_selected.drop(['co_cine_area_geral', 'co_cine_area_especifica', 'co_cine_area_detalhada', 'co_municipio'],axis=1)
```

```python
df_selected
```

```python
df_selected['no_curso'] = df['no_curso']
df_selected['no_municipio'] = df['no_municipio']
df_selected['no_cine_area_geral'] = df['no_cine_area_geral']
df_selected['tp_grau_academico'] = df['tp_grau_academico']
df_selected['enade_faixa'] = df['enade_faixa']
df_selected['cpc_faixa'] = df['cpc_faixa']
```

```python
df_selected
```

```python
df_selected.shape
```

```python
profile = ProfileReport(df_selected, title='Relatório de Dados Mínimo', minimal=True)
profile.to_file('cursos_ufrpe_3.html')
```

```python
df_selected.to_csv('cursos-ufrpe-pos-selecao-inicial.csv', sep=';', index=False)
```
