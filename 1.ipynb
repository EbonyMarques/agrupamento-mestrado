Aqui ocorrem:
1. a seleção dos cursos presenciais da UFRPE;
2. a "limpeza" nos nomes das variáveis e nos dados categóricos (trocando letras maiúsculas por minúsculas e removendo caracteres especiais);
3. a verificação da existência ou não de dados ausentes;
4. a adição dos indicadores dos cursos;

```python
import pandas as pd
from unidecode import unidecode
from pandas_profiling import ProfileReport
```

```python
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', None)
```

```python
df_2021 = pd.read_csv('../dados/MICRODADOS_CADASTRO_CURSOS_2021.CSV',encoding='ISO-8859-1',sep=';')
```

```python
df_indicadores = pd.read_csv('cursos_com_indicadores.csv', sep=',')
```

```python
# ufrpe tem codigo 587
# modalidade de ensino 1 eh presencial
df_2021 = df_2021.query('CO_IES == 587 and TP_MODALIDADE_ENSINO == 1').reset_index(drop=True).copy()
```

```python
df_2021.replace({'TP_GRAU_ACADEMICO':{1:'bacharelado',2:'licenciatura',3:'tecnologico',4:'bacharelado e licenciatura'}},inplace=True)
```

```python
def remove_acentos(texto):
    return unidecode(texto)
```

```python
def minusculo(texto):
    texto = texto.replace(' ', '_')
    return str(texto).lower()
```

```python
df_2021.columns = df_2021.columns.str.lower()
```

```python
colunas_categoricas = df_2021.select_dtypes(include=['object']).columns
for coluna in colunas_categoricas:
    df_2021[coluna] = df_2021[coluna].apply(remove_acentos)
    df_2021[coluna] = df_2021[coluna].apply(minusculo)
```

```python
df_2021
```

```python
df_indicadores.rename({'cod_inep':'co_curso'},axis=1, inplace=True)
```

```python
df_indicadores.co_curso = df_indicadores.co_curso.astype('int64')
```

```python
df_indicadores.columns
```

```python
# fazendo o merge do df de cursos com o df de indicadores dos cursos
resultado = pd.merge(df_2021, df_indicadores[['co_curso', 'enade_faixa', 'cpc_faixa']], on='co_curso', how='left')
```
